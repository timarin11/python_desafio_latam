## <font color='green'>Componentes de una funci칩n</font>

Python cuenta con funciones ya definidas -conocidas como **built-in functions**, que memos utilizado durante este m칩dulo como **print**, **input**, etc. Veremos ahora c칩mo podemos definir una funci칩n nosotros mismos (**user defined functions**) para realizar m치s f치cilmente las tareas que deseemos.

Supongamos que en un curso se realizan 3 pruebas y 5 tareas, adem치s de un examen final. La nota de cada estudiante se calcula considerando que:

- El promedio de las pruebas corresponde al 50% de la nota final
- El promedio de las tareas corresponde al 30% de la nota final
- El examen corresponde al 20% de la nota final.

La nota de cada estudiante depende, en total, de 12 valores: las 9 notas y los tres porcentajes o ponderaciones dadas. Si queremos establecer una f칩rmula que calcule la nota final tendremos que considerar estos 12 valores o **par치metros**. Pero ya que consideramos los porcentajes como valores fijos (aunque podr칤an modificarse), las notas obtenidas corresponder치n a los **argumentos** de nuestra funci칩n, es decir, los valores que recibiremos de un usuario para realizar nuestros c치lculos.

Considerando esto, analizaremos el siguiente bloque de c칩digo:

```python
def nota_final(prueba1,prueba2,prueba3,tarea1,tarea2,tarea3,tarea4,tarea5,examen):
    promedio_pruebas=round((prueba1+prueba2+prueba3)/3,1)
    promedio_tareas=round((tarea1+tarea2+tarea3+tarea4+tarea5)/5,1)
    final=round(0.5*promedio_pruebas+0.3*promedio_tareas+0.2*examen,1)
    print(final)
```

Podemos observar que:
- Utilizamos la palabra reservada **def** para se침alar que estamos definiendo una funci칩n.
- Luego viene el nombre de la funci칩n, en nuestro caso **nota_final**. Aunque pod콋iamos ponerle cualquier nombre, es una buena pr치ctica llamarla de una manera intuitiva, que se relacione con lo que realiza. Si utilizamos m치s de una palabra, se utiliza el gui칩n bajo para unirlas, y siempre con min칰sculas. Esto se conoce como **snake case**, y es un formato establecido para la creaci칩n de funciones en Python.
- Junto al nombre de la funci칩n, entre par칠ntesis, escribimos todos sus argumentos, y finalizamos con **:**.
- Luego de los **:**, si presionamos **enter**, el bloque de c칩digo estar치 con indentaci칩n, para delimitar que esto corresponde a la definici칩n o **cuerpo** de la funci칩n. Aqu칤 escribimos tomas las instrucciones que queremos que nuestra funci칩n ejecute.

Para utilizar nuestra funci칩n, simplemente la **invocamos** escribiendo entre los par칠ntesis los argumentos separados por comas. Por ejemplo:

```python
nota_final(3.5,4.2,5.7,4.2,5.5,4,7,4.5,4.3)
```

## <font color='green'>Retornos</font>

En los casos anteriores, la 칰ltima instrucci칩n de nuestras funciones ha sido imprimir un resultado. Esto puede resultarnos 칰til para ver ese resultado, pero no nos permite guardar este valor. Una alternativa a esto es definir un **retorno** para nuestra funci칩n, es decir, lo que queremos que retorne para almacenarlo, por ejemplo, en una variable. Por ejemplo,modificaremos la funci칩n anterior:

```python
def promedio_lista(lista):
    suma=0
    for i in lista:
        suma = suma + i
    promedio = round(suma/len(lista),1)
    return(promedio)

a = promedio_lista([5,4.7,6.5])
b = promedio_lista([5.5,6.2,4.5])
```

En esta ocasi칩n no obtenemos un resultado en pantalla, pero hemos almacenado estos resultados en las variables **a** y **b**, lo que nos permite imprimirlas e incluso operar entre ellas

```python
c = a + b
print(a)
print(b)
print(c)
```

Python nos ofrece, adem치s, la posibilidad de obtener m치s de un retorno para una funci칩n, lo que entrega como una tupla. Si nos interesa recuperar uno de sus elementos, simplemente podemos utilizar el 칤ndice correspondiente

```python
def promedio_lista(lista):
    suma=0
    for i in lista:
        suma = suma + i
    promedio = round(suma/len(lista),1)
    if promedio >= 4:
        resultado = 'aprobado'
    else:
        resultado = 'reprobado'
    return(promedio, resultado)

print(promedio_lista([5,4.7,6.5]))
print(promedio_lista([5,4.7,6.5])[1])
```

En ocasiones, hacer esto puede parecer m치s complejo ya que nos hace definir m치s funciones, pero a la larga, si tenemos que utilizar muchas veces un c치lculo determinado puede terminar resultando m치s 칰til rear algunas herramientas y luego utilizarlas, lo que nos permitir치 obtener c칩digos m치s limpios.

 ## <font color='green'>Variables locales y globales</font>

 Observa el siguiente bloque de c칩digo:

```python
t = 18

def sumat(a):
    g = a + t
    r = 2*g
    return(r)

sumat(8)
```

Observa que al principio de nuestro bloque hemos definido la variable **t = 18**, y luego hemos utilizado este valor dentro de una funci칩n. A la variable **t** se le ha asignado un valor **global**, es decir, fuera del entorno de una funci칩n, por lo que es posible acceder a ella desde cualquier parte del programa.

Compara lo que ocurre si imprimimos el valor de la variable g, que fue utilizada en la funci칩n **sumat**

```python
print(g)
```

La variable **g** es una **variable local**, puesto que su 치mbito solo corresponde al de la funci칩n en la que est치 definida. Si queremos que una variable local, definida en una funci칩n, pueda salir al 치mbito global, tendr치 que hacerlo mediante el comando **return**.

Observa que si utilizamos el nombre de una variable global en el 치mbito de una funci칩n, el valor cambiar치 dentro de la funci칩n **pero volver치 al valor definido fuera de ella**. Observa, por ejemplo, el siguiente c칩digo:

```python
def funcion_prueba(s):
    t = s + 5
    print(t)

funcion_prueba(3)
print(t)
```

En el primer caso, la instrucci칩n **print** se encuentra dentro de la funci칩n, por lo que utiliza el valor de **t** correspondiente. El segundo print est치 fuera de la funci칩n, por lo que utiliza el valor global de **t**.

En ocasiones, puede ser necesario modificar el valor de una variable global **dentro** de una funci칩n, aunque no se considera una buena pr치ctica ya que las variables globales suelen ser definidas as칤 con un objetivo. Para hacer esta modificaci칩n es preciso indicar dentro de la funci칩n que se est치 asignando un valor a la variable global, como se muestra.

```python
def modifica(f):
    global t
    t = f + 1

print(t)
modifica(2)
print(t)
```

 # <font color='green'>Funciones y DataFrames</font>

Anteriormente hab칤amos visto que el m칠todo **df.apply()** nos permit칤a aplicar **funciones lambda** a cada elemento de un DataFrame. En general, podemos aplicar una funci칩n cualquiera definida en Python a un DataFrame como podemos ver en los siguientes ejemplos:

- Aplicamos una funci칩n a todo un DataFrame

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})


def square(x):
    return x * x


df1 = df.apply(square)

print(df)
print(df1)
```

- Aplicamos una funci칩n por columna.

```python
import numpy as np

df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})

df1 = df.apply(np.sum, axis=0)
print(df1)

df1 = df.apply(np.sum, axis=1)
print(df1)
```

- Aplicamos por elemento, con applymap

```python
import math

df = pd.DataFrame({'A': [1, 4], 'B': [100, 400]})
df1 = df.applymap(math.sqrt)

print(df)
print(df1)
```

Tambi칠n podemos hacerlo para manipular strings.

```python
df = pd.DataFrame({'Name': ['Pankaj', 'Meghna'], 'Role': ['ceo', 'cto']})

df1 = df.applymap(str.upper)

print(df)
print(df1)
```
